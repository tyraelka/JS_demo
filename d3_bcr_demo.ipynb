# Bar Chart Race, Explained

* Code Source: https://observablehq.com/@d3/bar-chart-race-explained

```javascript
const fs = require("fs");
const { parse } = require("csv-parse");
const d3 = require("./nodejs/d3.js");
```

```javascript
data = []
fs.createReadStream("./data/category-brands.csv")
    .pipe(parse({delimiter: ",", from_line: 2}))
    .on("data", function (row){
        data.push(
            {
                "date": row[0],
                "name": row[1],
                "category": row[2],
                "value": row[3]
            }
        );
    });
```

```javascript
n = 12
```

```javascript
names = new Set(data.map(d => d.name))
```

```javascript
datevalues = Array.from(d3.rollup(data, ([d]) => d.value, d => d.date, d => d.name))
    .map(([date, data]) => [new Date(date), data])
    .sort(([a], [b]) => d3.ascending(a, b))
```

```javascript
function rank(value) {
    const data = Array.from(names, name => ({name, value: value(name)}));
    data.sort((a, b) => d3.descending(a.value, b.value));
    for (let i = 0; i < data.length; i++) data[i].rank = Math.min(n, i);
    return data;
}
```

```javascript
rank(name => datevalues[0][1].get(name))
```

```javascript
k = 10
```

```javascript
keyframes = {
    const keyframes = [];
    let ka, a, kb, b;
    for ([[ka, a], [kb, b]] of d3.pairs(datevalues)) {
        for (let i = 0; i < k; i++) {
            const t = i / k;
            keyframes.push([
            new Date(ka * (1 - t) + kb * t),
            rank(name => (a.get(name) || 0) * (1 - t) + (b.get(name) || 0) * t)
            ]);
        }
    }
    keyframes.push([new Date(kb), rank(name => b.get(name) || 0)]);
    return keyframes;
}
```
